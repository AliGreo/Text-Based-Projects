```python
import numpy as np

dk = 4
num_words = 6

# create the used embedding matrix
embeddings = np.random.normal(loc=0, scale=1, size=(num_words, dk))

# create the (QKV) weight matrices
qw = np.random.uniform(low=-1, high=1, size=(dk,dk))
kw = np.random.uniform(low=-1, high=1, size=(dk,dk))
vw = np.random.uniform(low=-1, high=1, size=(dk,dk))

# compute the QKV
q = embeddings @ qw
k = embeddings @ kw
v = embeddings @ vw

def softmax(x):
    """ a score function to output a probabilaty for each word poistion"""
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

# compute the attention weights
attention_scores = softmax((q @ k.T) / np.sqrt(dk))
print(f"Attention weights:\n{attention_scores}\n")
# compute the attention respresentation
attention_representation = attention_scores @ v
print(f"Attention Respresentation:\n{attention_representation}")
```
